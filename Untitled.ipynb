```python
import random
import json
import nltk
from nltk.stem import WordNetLemmatizer
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD

nltk.download('punkt')
nltk.download('wordnet')
```

```python

# load the intents file
with open('hengameh_intents.json') as file:
    data = json.load(file)
```

```python
# initialize lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# create an empty list to store words
words = []
# create an empty list to store classes
classes = []
# create an empty list to store patterns
patterns = []
# create a dictionary to store responses
responses = {}
```

```python
# loop through each intent in the intents file
for intent in data['intents']:
    # loop through each pattern in the intent
    for pattern in intent['patterns']:
        # tokenize each pattern
        w = nltk.word_tokenize(pattern)
        # add the tokenized words to the words list
        words.extend(w)
        # add the tokenized words and the intent tag to the patterns list
        patterns.append((w, intent['tag']))
    # add the intent tag to the classes list
    if intent['tag'] not in classes:
        classes.append(intent['tag'])
    # add the responses to the responses dictionary
    responses[intent['tag']] = intent['responses']
```

```python
print(words)
print('\n2\n',classes)
print('\n3\n',patterns)
print('\n4\n',responses)
```

```python
# lemmatize and lowercase each word in the words list
words = [lemmatizer.lemmatize(w.lower()) for w in words if w != '?']
words
```

```python
# sort the words list and remove duplicates
words = sorted(list(set(words)))
words
```

```python
# sort the classes list
classes = sorted(list(set(classes)))
classes
```

```python
# create training data
training = []
output_empty = [0] * len(classes)
```

```python
for pattern in patterns:
    bag = []
    pattern_words = pattern[0]
    pattern_words = [lemmatizer.lemmatize(word.lower()) for word in pattern_words]
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)
    output_row = list(output_empty)
    output_row[classes.index(pattern[1])] = 1
    training.append([bag, output_row])
```

```python
output_row
```

```python
training
```

```python
# shuffle the training data
random.shuffle(training)
training = np.array(training)
```

```python
# split the training data into input and output
train_x = list(training[:, 0])
train_y = list(training[:, 1])
```

```python
train_x
```

```python
# build the model
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))
```

```python
# compile the model
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
# fit the model to the training data
hist = model.fit(np.array(train_x), np.array(train_y), epochs=200, batch_size=5, verbose=1)

# save the trained model
model.save('chatbot_model.h5', hist)
```
