```python
import numpy as np
import matplotlib.pyplot as plt

def f_prime(x):
    numerator = 2*np.sin(2)*(np.sin(2*x) - np.sin(2)) + 2*np.cos(2)*(np.cos(2) - np.cos(2*x))
    denominator = (np.sin(2*x) - np.sin(2))**2 + (np.cos(2) - np.cos(2*x))**2
    return -numerator/denominator

x = np.linspace(-2, 0.5, 1000)
y = f_prime(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Plot of f'(x)")
plt.grid()
plt.show()
```
